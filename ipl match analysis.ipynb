*Total match played by ROHIT SHARMA in IPL*

```python
def total_match_of_rohitsharma(df):
  match_no=[]
  for index in df.index:
   if (df["Team1"][index]=='Mumbai Indians')&('RG Sharma' in df["Team1Players"][index]):

    match_no.append(index)
   elif (df["Team2"][index]=='Mumbai Indians')&('RG Sharma' in df["Team2Players"][index]):

    match_no.append(index)
  return match_no
```

```python
x=total_match_of_rohitsharma(df)
len(x)
```
